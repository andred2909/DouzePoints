# Lingua

Un paragrafo molto breve è da dedicare alla lingua: quello che è un contest nato sull'onda sanremese negli anni '50 del Novecento, vede susseguirsi da allora, generalmente, ogni anno una presentazione trilingue: la conduzione, infatti, viene effettuata in lingua inglese, francese e nella lingua tipica del paese che ospita l'evento.

Valutiamo, quindi, se lo stesso si può dire per il numero di lingue utilizzate nelle canzoni che analizziamo in questo dataset, sia tenendo in considerazione tutte le partecipanti, che solo tra le vincitrici:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Carica il file CSV
file_path = './song_data.csv'
pd.set_option('display.max_rows', None)
data = pd.read_csv(file_path, encoding='latin1')

language_counts = data['language'].value_counts()

for language, count in language_counts.items():
    print(f"{language}: {count}")

# Visualizza la frequenza delle lingue
plt.figure(figsize=(12, 8))
sns.barplot(x=language_counts.values, y=language_counts.index, palette='viridis')
plt.title('Frequenza delle Lingue tra le Canzoni Partecipanti all\'Eurovision (2009-2016)')
plt.xlabel('Numero di canzoni')
plt.ylabel('Lingua')
plt.show()

# Filtra le canzoni vincenti
winning_songs = data[data['final_place'] == 1]

# Conta la frequenza delle lingue tra le canzoni vincenti
language_counts = winning_songs['language'].value_counts()

for language, count in language_counts.items():
    print(f"{language}: {count}")

# Visualizza la frequenza delle lingue
plt.figure(figsize=(12, 8))
sns.barplot(x=language_counts.values, y=language_counts.index, palette='viridis')
plt.title('Frequenza delle Lingue tra le Canzoni Vincenti all\'Eurovision')
plt.xlabel('Numero di Vittorie')
plt.ylabel('Lingua')
plt.show()
```

Come si può notare, è piuttosto marcata la differenza tra le due frequenze.
C'è, però da tenere a mente un paio di cose:

* In primis dobbiamo notare che, per settaggio del dataset stesso, molte canzoni hanno utilizzato, all'interno dello stesso brano, più lingue e, soprattutto più dialetti.
    * Ciò significa che, molto spesso, canzoni che trattano più di una lingua nello stesso brano, vengono considerate (anche ai fini statistici della stessa European Broadcast Union) come "lingue separate"
    * Sulla stessa onda del discorso dei dialetti, questo in realtà spiega anche la motivazione per cui, tra le vincitrici risultano separate a livello linguistico la vittoria a Torino dell'Ucraina con il brano *Stefania* dei *Kalush Orchestra* nel 2022 e la vittoria a Stoccolma, nel 2016, della stessa Ucraina con il brano *1944* di *Jamala*.

In generale, però, rimane notevole l'utilizzo della lingua inglese come mezzo di comunicazione universale: ben 383 brani partecipanti (di cui 10 vincitori) hanno deciso, a prescindere dalla propria lingua di origine, di esporre la propria canzone in lingua inglese.

Note di merito, però, vanno anche al *portoghese* e all'*italiano*: uniche due lingue presenti con una frequenza elevata sia nella partecipazione al contest in generale, che nella vittoria del contest stesso.
